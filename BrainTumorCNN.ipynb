```python
import glob
import os
import numpy as np
import pandas as pan
import matplotlib.pyplot as plotter
import seaborn as sns
import tensorflow as tf
from sklearn.model_selection import train_test_split
from keras.layers import Dense, Conv2D, MaxPool2D, Flatten
from keras.optimizers import Adam
from keras.models import Sequential
from keras.callbacks import ReduceLROnPlateau, EarlyStopping
```

```python
PATH = "Data"
files = [i for i in glob.glob(PATH + "//*//*")]
np.random.shuffle(files)
labels = [os.path.dirname(i).split("/")[-1] for i in files]
data = zip(files, labels)
dataframe = pan.DataFrame(data, columns = ["Image", "Label"])
```

```python
sns.countplot(x = dataframe["Label"])
plotter.xticks(rotation = 90)
```

```python
print(dataframe["Label"].iloc[0])
array = plotter.imread(dataframe["Image"].iloc[15])
plotter.imshow(array)
```

```python
mappings = {k: i for i, k in enumerate(dataframe["Label"].unique())}

features = dataframe["Image"].values 
label = dataframe["Label"].map(mappings).values
print(mappings)

feature_train, feature_test, label_train, label_test = train_test_split(features, label, test_size = 0.2, random_state = 42)
train_dataset = tf.data.Dataset.from_tensor_slices((feature_train, label_train))
test_dataset = tf.data.Dataset.from_tensor_slices((feature_test, label_test))
                                                   
def preprocessor(image, label):
    image = tf.io.read_file(image)  
    image = tf.io.decode_image(image, channels = 1, expand_animations = False)
    image = tf.image.resize(image, (256, 256))
    return image/255, tf.one_hot(label, depth = 4)  

train_dataset = train_dataset.map(preprocessor)
test_dataset = test_dataset.map(preprocessor)

train_dataset_loader = train_dataset.batch(128).shuffle(64).prefetch(64)
test_dataset_loader = test_dataset.batch(128).shuffle(128).prefetch(128)
```

```python
callback_one = EarlyStopping(patience = 4, restore_best_weights = True)
callback_two = ReduceLROnPlateau(patience = 2)

image_learner = Sequential()
image_learner.add(Conv2D(32, (2,2 ), input_shape = (256, 256, 1), activation = "relu"))
image_learner.add(MaxPool2D((2, 2)))
image_learner.add(Conv2D(64, (2, 2), activation = "relu"))
image_learner.add(MaxPool2D((2, 2)))
image_learner.add(Conv2D(128, (2, 2), activation = "relu"))
image_learner.add(MaxPool2D((2, 2)))
image_learner.add(Conv2D(256, (2, 2), activation = "relu"))
image_learner.add(MaxPool2D((2, 2)))
image_learner.add(Flatten())
image_learner.add(Dense(128, activation = "relu"))
image_learner.add(Dense(4, activation = "softmax"))

opt = Adam()
image_learner.compile(loss = "categorical_crossentropy", optimizer = opt, metrics = ["accuracy"])
image_learner.summary()
```

```python
image_learner.fit(train_dataset_loader, validation_data = test_dataset_loader, callbacks = [callback_one, callback_two], epochs = 17)
```

```python
example_image_path = feature_test[0]
example_image = tf.io.read_file(example_image_path)
example_image = tf.io.decode_image(example_image, channels=1, expand_animations=False)
example_image = tf.image.resize(example_image, (256, 256))
example_image = example_image / 255
```

```python
prediction = image_learner.predict(tf.expand_dims(example_image, axis=0))
```

```python
predicted_class = np.argmax(prediction)
```

```python
predicted_label = [k for k, v in mappings.items() if v == predicted_class][0]

true_class = [k for k, v in mappings.items() if v == label_test[0]][0]
```

```python
import matplotlib.pyplot as plt

plt.imshow(example_image.numpy().reshape(256, 256), cmap='gray')
plt.title(f"Predicted Class: {predicted_label}, \nTrue Class: {true_class}")
plt.show()
```
